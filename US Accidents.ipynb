## Feature importance

```python
#Work to identify feature importance 
#start by creating the target value (severity) and the training x columns
train_y = df['Severity'].values
x_cols = [col for col in df.columns if col not in ['Severity'] if df[col].dtype=='float64'or df[col].dtype=='int64']
train_col= df[x_cols]

fearture_name = train_col.columns.values 

from sklearn import ensemble 
#using ensemble extratreesregressor here to fit a number of randomized decision trees on the model. In this case we use 25 trees
#each tree will have a max depth of 30 nodes 
#set max features to 0.3 for more randomness
#set n jobs to -1 to use all processors
#set random state to 0 to return the same results for each execution
model = ensemble.ExtraTreesRegressor(n_estimators=25, max_depth=30, max_features=0.3, n_jobs=-1, random_state=0)
model.fit(train_col,train_y)

#using the feature importances method in ensemble (computed as mean and standard dev of accumulation of impurity decrease within each tree)
importance = model.feature_importances_
std = np.std([tree.feature_importances_ for tree in model.estimators_],axis=0)
indices = np.argsort(importance)[::-1][:20]

#plot feature importances
plt.figure(figsize=(20,7))
plt.title("Feature importances")
plt.bar(range(len(indices)), importance[indices], color="r", yerr=std[indices], align="center")
plt.xticks(range(len(indices)), fearture_name[indices], rotation='vertical')
plt.xlim([-1, len(indices)])
plt.show()
```

# Train Test Split

```python
y = df['Severity'].copy()
X = df.drop('Severity', axis=1).copy()
```

```python
y.unique()
```

```python
y = y-1
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#using standard scaler to try and normalize the features before creating the models
#by doing this mean of each feature will be 0 and the standard dev of each feature will be 1
#we do this to reduce outliers and scales that could bring bias into the data

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=100)
```

```python
X_train.shape
```

# Models

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

# Model 1

```python
#will be used to define the model
inputs = tf.keras.Input(shape=(X.shape[1],))
#using rectified linear unit activation -- standard activation
x = tf.keras.layers.Dense(64, activation='relu')(inputs)
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
#softmax activation to convert vector of values to probability distribution
outputs = tf.keras.layers.Dense(4, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

#adam optimizer is a stochastic gradient descent method that is based on adaptive estimation of first-order and second-order moments
#well suited for problems that are large in terms of data/parameters
#computationally efficient and has little memory requirement

#use sparse categorical crossentropy loss function where there are two or more label classes

#accuracy predicts how often prediction equal labels
#metric creates two local variables, total and count that are used to compute the frequency of correct predictions
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

batch_size = 40
epochs = 100

#reduceLROnPlateau call back will reduce learning rate when a metric has stopped improving
#earlycallback will stop training when metric has stopped improving
history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau(),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=3,
            restore_best_weights=True
        )
    ]
)
```

```python
print("Test Accuracy:", model.evaluate(X_test, y_test, verbose=0)[1])
```

```python
losses=pd.DataFrame(history.history)
```

```python
l = pd.DataFrame({'loss':losses['loss'], 'validation_loss':losses['val_loss']})
```

## Model 2

```python
inputs = tf.keras.Input(shape=(X.shape[1],))
x = tf.keras.layers.Dense(64, activation='relu')(inputs)
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
outputs = tf.keras.layers.Dense(4, activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

batch_size = 32
epochs = 100

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    callbacks=[
        tf.keras.callbacks.ReduceLROnPlateau()
    ]
)
```

```python
print("Test Accuracy:", model.evaluate(X_test, y_test, verbose=0)[1])
```

```python
losses=pd.DataFrame(history.history)
l = pd.DataFrame({'loss':losses['loss'], 'validation_loss':losses['val_loss']})
```
